## Define CNN

```python
# Define the CNN model
import torch.nn as nn
import torch.nn.functional as F

class CNN(nn.Module):
    def __init__(self):
        super(CNN, self).__init__()
        # Define convolutional layers
        self.conv1 = nn.Conv2d(3, 6, 5, 1, 0)
        self.bn1=nn.BatchNorm2d(6)
        self.pool = nn.MaxPool2d(2, 2)
        self.conv2 = nn.Conv2d(6, 16, 3)
        self.bn2=nn.BatchNorm2d(16)
        self.dropout=nn.Dropout(0,2)

        # Define fully connected layers
        self.fc1 = nn.Linear(16 * 6 * 6, 144)
        self.bn3=nn.BatchNorm2d(144)
        self.fc2 = nn.Linear(144, 72)
        self.fc3 = nn.Linear(72, 10)

    def forward(self, x):
        # Apply convolutional and pooling layers
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        # Reshape the tensor for fully connected layers
        x = x.view(-1, 16 * 5 * 5)
        # Apply fully connected layers with ReLU activation function
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        # Apply the final fully connected layer
        x = self.fc3(x)
        return x

# Create an instance of the model
net = CNN()
```
